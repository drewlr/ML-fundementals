## ML Metrics from scratch
Let's explore some basic ML metrics, and implement a function that takes y_true and y_pred (binary) and returns precision, recall, F1, and accuracy. Handle division by zero.

- **Precision**: out of all predictions y_pred that predicted true, how many were true
- **Recall**: out of all predictions y_true that are true, how many were predicted
- **F1**: precision*recall / (precision + recall)
- **Accuracy**: % of y_pred that match y_true

```python
import numpy as np
```

```python
n=100
threshold = 0.4
y_true = np.random.randint(0, 2, size=n)
y_proba = np.random.rand(n)
y_pred = (y_proba >= threshold).astype(int)
```

```python
denom = np.sum(y_pred == 1)
precision = np.sum((y_pred == 1) & (y_true == 1)) / denom if denom > 0 else 0
precision
```

```python
denom = np.sum(y_true == 1)
recall = np.sum((y_pred == 1) & (y_true == 1)) / denom if denom > 0 else 0
recall
```

```python
f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
f1
```

```python
denom = len(y_true)
accuracy = np.sum(y_pred == y_true) / denom if denom > 0 else 0
accuracy
```

**Implement AUC-ROC from scratch**

given y_true and y_scores (probabilities), compute TPR and FPR at every threshold and return the area under the curve using the trapezoidal rule.

```python
def auc_roc(y_true, y_scores):
    # 1. Get all unique thresholds, sorted descending
    thresholds = np.sort(np.unique(y_scores))[::-1]

    tprs = []
    fprs = []

    total_pos = np.sum(y_true == 1)
    total_neg = np.sum(y_true == 0)

    for t in thresholds:
        y_pred = (y_scores >= t).astype(int)
        tp = np.sum((y_pred == 1) & (y_true == 1))
        fp = np.sum((y_pred == 1) & (y_true == 0))
        tprs.append(tp / total_pos if total_pos > 0 else 0)
        fprs.append(fp / total_neg if total_neg > 0 else 0)

    # 2. Add the origin point (0, 0)
    fprs = [0] + fprs
    tprs = [0] + tprs

    # 3. Area under curve using trapezoidal rule
    auc = 0
    for i in range(1, len(fprs)):
        auc += (fprs[i] - fprs[i-1]) * (tprs[i] + tprs[i-1]) / 2

    return auc

auc_roc(y_true, y_proba)
```
