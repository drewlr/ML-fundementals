```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's create some random data

```python
X = np.array([32.50234527, 53.42680403, 61.53035803, 47.47563963, 59.81320787,
       55.14218841, 52.21179669, 39.29956669, 48.10504169, 52.55001444,
       45.41973014, 54.35163488, 44.1640495 , 58.16847072, 56.72720806,
       48.95588857, 44.68719623, 60.29732685, 45.61864377, 38.81681754])
Y = np.array([31.70700585, 68.77759598, 62.5623823 , 71.54663223, 87.23092513,
       78.21151827, 79.64197305, 59.17148932, 75.3312423 , 71.30087989,
       55.16567715, 82.47884676, 62.00892325, 75.39287043, 81.43619216,
       60.72360244, 82.89250373, 97.37989686, 48.84715332, 56.87721319])
```

Let's plot the data

```python
plt.figure(figsize = (5,2))
plt.scatter(X, Y, marker='o', color='red')
plt.plot(min(X), max(X), color='blue',markerfacecolor='red',
         markersize=10,linestyle='dashed')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

Our prediction model is
y_pred = weight*x + bias

our cost function is

cost = sum( (y_true - y_pred)^2 ) / n
cost = sum( (y_true - weight*x + bias)^2) / n

To optimize the prediction model, we iteratively adjust the weight weight and bias, by learning from the error of each iteration.
e.g. we start with some random weights and look at the prediction. If we on average we overshot, we tune the weights so that the predictions are smaller. We check new predictions, and we keep going. To adjust the weights, we look at how the cost function changes as the weight change - i.e. we look at the derivative of the cost function w.r.t. weights:

gradient_weight = -2*sum( x* (y_true - weight*x + bias)) / n = -2*sum(x* (y_true - y_pred))/n

gradient_bias = -2/ * sum(y_true - y_pred)/n


and we adjust the weights in the direction that minimizes the error (i.e. the cost function)

Let's initialize some parameters

```python
current_weight = 0.1
current_bias = 0.01
iterations = 5000
learning_rate = 0.01
n = float(len(X))
y_predicted = Y + 0.01
costs = []
weights = []
previous_cost = None
x = X
y = Y
learning_rate = 0.0001
stopping_threshold = 1e-6
```

Let's write the core code

```python
for i in range(iterations):
        
        # Making predictions
        y_predicted = (current_weight * x) + current_bias
        
        # Calculating the current cost
        squared_error = np.power(y -y_predicted,2)
        current_cost = np.sum(squared_error)/len(squared_error)

        # If the change in cost is less than or equal to 
        # stopping_threshold we stop the gradient descent
        if previous_cost and abs(previous_cost-current_cost)<=stopping_threshold:
            break
        
        previous_cost = current_cost

        costs.append(current_cost)
        weights.append(current_weight)
        
        # Calculating the gradients
        weight_derivative = -(2/n) * sum(x * (y-y_predicted))
        bias_derivative = -(2/n) * sum(y-y_predicted)
        
        # Updating weights and bias
        current_weight = current_weight - (learning_rate * weight_derivative)
        current_bias = current_bias - (learning_rate * bias_derivative)
                
        # Printing the parameters for each 1000th iteration
        print(f"Iteration {i+1}: Cost {current_cost}, Weight \
        {current_weight}, Bias {current_bias}")
```

Let's plot the results

```python
# Visualizing the weights and cost at for all iterations
plt.figure(figsize = (8,6))
plt.plot(weights, costs)
plt.scatter(weights, costs, marker='o', color='red')
plt.title("Cost vs Weights")
plt.ylabel("Cost")
plt.xlabel("Weight")
plt.show()
```

```python
# Making predictions using estimated parameters
Y_pred = current_weight*X + current_bias

# Plotting the regression line
plt.figure(figsize = (8,6))
plt.scatter(X, Y, marker='o', color='red')
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='blue',markerfacecolor='red',
         markersize=10,linestyle='dashed')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
